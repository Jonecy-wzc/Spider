### Q4: Now the scraping. Please scrape the information from the webpage, with the following requirements.

#### 1. You are expected to scrape a set of information from the web, i.e., elements inthe tables, elements in several tags, serveral files, or any other information that fit your purpose;

#### 2. You are expected to use at least one of the control flow statements (i.e., conditional statements, loops, and functional calls) to fine tune the code.

#### 3. You are expected to store the information in a dataframe format, and output the file in CSV format.

```python
# Functional calls
def parse(div):
    # save all the information in a dict
    info = dict()
    paper_title = div.find("h5", class_ = "issue-item__title").find("a").text
    # Loops
    author_list = []
    for tag_a in div.find("ul").find_all("a"):
        author_list.append(tag_a.text.strip())
    publish_time = div.find("div", class_ = "rlist--inline separator toc-item__detail").find("p").text.split(":")[-1]
    abstract = div.find("span", class_ = "hlFld-Abstract").text
    url = div.find_all("p")[-1].find("a")["href"]
    # load data into the blank dict
    info["Title"] = paper_title
    info["Authors"] = ",".join(author_list)
    info["PubTime"] = publish_time
    info["Abstract"] = abstract
    info["URL"] = url 
    return info

# Functional calls
def get_all_information(soup, k):
    # save all information 
    info_list = []
    # Loops 
    for idx, div in enumerate(soup.find_all("div", class_ = "issue-item")):
        # scrape paper information 
        info = parse(div)
        info_list.append(info)
        # Conditional statements
        # we only need k items, if the number of items is greater than k, stop and return
        if idx+1 >= k:
            break
    return info_list

# Functional calls
def save_to_file(info_list, outPath):
    # transform to dataframe
    output = pd.DataFrame(info_list) 
    output.to_csv(outPath, index = False)
    return  
```

```python
# Given the initial value
k = 340
outPath = "../assignment01_YUJiachi_21408424/papers.csv"
# start
info_list = get_all_information(soup, k)
save_to_file(info_list, outPath)
```
